```python
import numpy as np
from matplotlib import pyplot as plt



signal = np.zeros(101)+ 0.05*np.random.rand(101)
time = [x for x in range(len(signal))]

fig, ax = plt.subplots(figsize=(12, 6))

signal[4]=0.6;
signal[5]=0.96;
signal[6]=0.6;

plt.rcParams.update({'font.size': 20})
plt.plot(time, signal, '-',linewidth=3)
plt.ylim([0, 1])
#plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.xlabel("Zeit t")
plt.xticks([])
plt.ylabel("Amplitude (normiert)")
#plt.title("Ultraschall-Echomessung")

plt.savefig('B.svg')
plt.show()
```
